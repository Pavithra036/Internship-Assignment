### Storytelling with Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv(r"C:\Users\PAVITHRA H R\Desktop\Internship\Private_data.csv")

# ---- 1. Bar Chart (Top 10 Students by Total Score) ----
plt.figure()
top_students = df.sort_values(by='TOTAL SCORE (OUT of 100)', ascending=False).head(10)
plt.bar(top_students['NAME OF THE STUDENT'], top_students['TOTAL SCORE (OUT of 100)'])
plt.xticks(rotation=45)
plt.title("Top 10 Students by Total Score")
plt.xlabel("Student Name")
plt.ylabel("Total Score")
plt.tight_layout()
plt.show()

# ---- 2. Pie Chart (Domain Distribution) ----
plt.figure()
df['Domain'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top Domains Distribution")
plt.ylabel("")
plt.show()

# ---- 3. Histogram (Total Score Distribution) ----
plt.figure()
df['TOTAL SCORE (OUT of 100)'].plot(kind='hist', bins=20)
plt.title("Distribution of Total Scores")
plt.xlabel("Total Score")
plt.ylabel("Frequency")
plt.show()
```
