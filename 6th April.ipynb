```python
import cv2
import numpy as np

# Load image
image = cv2.imread("image.jpg")   # keep image in same folder

# Print shape
print("Image Shape:", image.shape)

# Print pixel values (first 5 pixels)
print("\nPixel Values (first 5):")
print(image[0][0:5])

# Print channels
print("\nNumber of Channels:", image.shape[2])

# Split channels
b, g, r = cv2.split(image)

print("\nBlue channel sample:", b[0][0:5])
print("Green channel sample:", g[0][0:5])
print("Red channel sample:", r[0][0:5])
```
