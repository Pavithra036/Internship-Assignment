### NLP Mini App (Chatbot)

```python
import random

# Predefined responses (intent-based)
responses = {
    "hello": ["Hi there!", "Hello!", "Hey! How can I help you?"],
    "how are you": ["I'm fine!", "Doing great!", "All good 😊"],
    "your name": ["I'm an NLP chatbot.", "You can call me ChatBot."],
    "bye": ["Goodbye!", "See you later!", "Bye! Have a nice day!"],
    "default": ["Sorry, I don't understand.", "Can you rephrase that?"]
}

# Chat function
def chatbot():
    print("Chatbot: Hello! Type 'bye' to exit.")

    while True:
        user_input = input("You: ").lower()

        if "bye" in user_input:
            print("Chatbot:", random.choice(responses["bye"]))
            break

        found = False

        for key in responses:
            if key in user_input:
                print("Chatbot:", random.choice(responses[key]))
                found = True
                break

        if not found:
            print("Chatbot:", random.choice(responses["default"]))

# Run chatbot
chatbot()
```
