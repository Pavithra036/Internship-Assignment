## House Price Predictor

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# ---- 1. Create Dataset ----
data = {
    "Area": [500, 800, 1000, 1200, 1500, 1800],  # sqft
    "Price": [50, 80, 100, 120, 150, 180]       # Lakhs
}

df = pd.DataFrame(data)

print("\n--- Dataset ---")
print(df)

# ---- 2. Features & Labels ----
X = df[["Area"]]     # Feature
y = df["Price"]     # Label

# ---- 3. Train Model ----
model = LinearRegression()
model.fit(X, y)

# ---- 4. Prediction ----
new_area = [[2000]]   # New house area
predicted_price = model.predict(new_area)

print("\nPredicted price for 2000 sqft house:", predicted_price[0], "Lakhs")
```
