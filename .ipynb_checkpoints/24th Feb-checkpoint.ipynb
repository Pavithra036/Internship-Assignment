
### Dataset Detective

```python
import pandas as pd

# Load dataset using raw string (important!)
df = pd.read_csv(r"C:\Users\PAVITHRA H R\Desktop\Internship\Private_data.csv")

# Top rows
print("\n--- Top 5 Rows ---")
print(df.head())

# Highest value column (based on mean)
numeric_cols = df.select_dtypes(include=['number'])
highest_col = numeric_cols.mean().idxmax()
print("\nHighest value column:", highest_col)

# Missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())
```
