### Word Importance Explorer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
#  1. Sample Documents ----
documents = [
    "Machine learning is very useful",
    "Data science uses machine learning",
    "Python is used in data science",
    "Machine learning and AI are trending",
    "Data analysis is important in science"
]
```

```python
# 2. Apply TF-IDF ----
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# Convert to DataFrame
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print("\n--- TF-IDF Matrix ---")
print(df)
```

```python
#  3. Get Top Keywords for each document ----
print("\n--- Top Keywords ---")
for i, row in df.iterrows():
    top_words = row.sort_values(ascending=False).head(3)
    print(f"Document {i+1}: {list(top_words.index)}")
```
