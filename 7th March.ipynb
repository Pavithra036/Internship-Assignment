## KNN in Real Life

```python
'''
1. Platforms like Netflix use recommendation systems to suggest movies/shows.
2. KNN (K-Nearest Neighbors) works by finding users with similar preferences.
3. If two users like similar movies, then the system recommends movies liked by one user to the other.

'''
```

```python
# user data
'''
User  Action  Comedy  Romance
 A      5       3       2
 B      4       2       1 
 C      1       5       4

'''
```

```python
# EXAMPLE 

import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

# User preferences
A = np.array([[5, 3, 2]])
B = np.array([[4, 2, 1]])
C = np.array([[1, 5, 4]])

# Calculate distances
dist_AB = euclidean_distances(A, B)
dist_AC = euclidean_distances(A, C)

print("Distance between A and B:", dist_AB[0][0])
print("Distance between A and C:", dist_AC[0][0])

# Find closest user
if dist_AB < dist_AC:
    print("User B is most similar to User A")
else:
    print("User C is most similar to User A")
```

```python
'''
KNN helps in recommendation systems by finding similar users based on preferences.
It is simple, effective, and widely used in platforms like Netflix for personalized suggestions.

'''
```
