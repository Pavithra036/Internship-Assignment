##  Build Your First Dataset

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

#  1. Create Dataset 
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [35, 40, 50, 55, 65, 70, 80, 90]
}

df = pd.DataFrame(data)

print("\n--- Dataset ---")
print(df)

# 2. Features & Labels 
X = df[["Study Hours"]]   # Feature (input)
y = df["Marks"]           # Label (output)

# 3. Train Model 
model = LinearRegression()
model.fit(X, y)

# 4. Prediction 
predicted = model.predict([[9]])
print("\nPredicted marks for 9 study hours:", predicted[0])
```
